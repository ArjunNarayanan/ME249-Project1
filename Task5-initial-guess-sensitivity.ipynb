```python
import matplotlib.pyplot as plt

ni1 = [1,10,100,1000]
ni1err = [0.03,0.0266,0.024,0.85]

ni2 = [3,4,5,6,7,8,9,10]
ni2err = [0.027,0.031,0.031,0.03,0.03,0.99,1.38,2.09]

ni3 = [0.35,1,10,100,1000]
ni3err = [0.027,0.033,0.04,0.047,0.047]
```

```python
fig,ax = plt.subplots()
ax.plot(ni1,ni1err,"-o")
ax.set_xscale("log")
ax.set_xlabel("Initial guess ni1")
ax.set_ylabel("Minimum average error aFerrMeanMin")
ax.set_title("Mean minimum error vs. initial guess for parameter n1")
fig.savefig("mean-error-vs-ni1.png")
```
